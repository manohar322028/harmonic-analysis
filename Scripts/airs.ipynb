```python
import os, re, glob
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.animation as mpl_anim
from matplotlib.ticker import AutoMinorLocator
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# ── point matplotlib to the conda-installed ffmpeg ─────────────────────────
matplotlib.rcParams["animation.ffmpeg_path"] = (
    r"C:\Users\ACER\anaconda3\envs\gis-env\Library\bin\ffmpeg.exe"
)

# ── config (same bounding box as all previous cells) ───────────────────────
lon_min, lon_max = -107.5, -85.0
lat_min, lat_max =   24.5,  37.5

VARIABLES = [
    "co2_vmr_uppertrop",
    "co_mmr_midtrop",
    "ch4_mmr_midtrop",
    "o3_tot",
    "h2o_vap_tot",
    "surf_air_temp",
]

meta = {
    "co2_vmr_uppertrop": dict(
        long_name="CO₂ Upper Troposphere VMR", disp_unit="ppm",
        scale=1e6, offset=0.0,
        line_color="#c0392b", fill_color="#f5cba7", cmap="YlOrRd",
    ),
    "co_mmr_midtrop": dict(
        long_name="CO Mid-Troposphere MMR", disp_unit="ppb",
        scale=1e9, offset=0.0,
        line_color="#d35400", fill_color="#f9e79f", cmap="YlOrBr",
    ),
    "ch4_mmr_midtrop": dict(
        long_name="CH₄ Mid-Troposphere MMR", disp_unit="ppb",
        scale=1e9, offset=0.0,
        line_color="#1e8449", fill_color="#a9dfbf", cmap="YlGn",
    ),
    "o3_tot": dict(
        long_name="Total Column Ozone", disp_unit="kg m⁻²",
        scale=1.0, offset=0.0,
        line_color="#1a5276", fill_color="#aed6f1", cmap="PuBu",
    ),
    "h2o_vap_tot": dict(
        long_name="Total Column Water Vapour", disp_unit="kg m⁻²",
        scale=1.0, offset=0.0,
        line_color="#6c3483", fill_color="#d7bde2", cmap="BuPu",
    ),
    "surf_air_temp": dict(
        long_name="Near-Surface Air Temperature", disp_unit="°C",
        scale=1.0, offset=-273.15,
        line_color="#117a8b", fill_color="#a2d9ce", cmap="RdYlBu_r",
    ),
}

DATA_DIR = r"D:\COCCON\Work\Data\SNDRAQIL3SMCCP_2"
SAVE_DIR = r"D:\COCCON\Work\Figures"
os.makedirs(SAVE_DIR, exist_ok=True)

# ── Phase 1: load every file once – spatial maps + scalar means ────────────
files    = sorted(glob.glob(os.path.join(DATA_DIR, "*.nc")))
date_pat = re.compile(r"\.(\d{8})\.")

spatial  = {v: [] for v in VARIABLES}   # list of 2-D arrays  (lat × lon)
ts_vals  = {v: [] for v in VARIABLES}   # list of scalar spatial means
dates    = []
lats = lons = None

print(f"Loading data from {len(files)} files …")
for f in files:
    hit = date_pat.search(os.path.basename(f))
    if not hit:
        continue
    date = pd.Timestamp(hit.group(1))
    if date.year > 2025:
        continue
    with xr.open_dataset(f, decode_timedelta=False) as ds:
        if lats is None:
            lats = ds["lat"].sel(lat=slice(lat_min, lat_max)).values
            lons = ds["lon"].sel(lon=slice(lon_min, lon_max)).values
        for var in VARIABLES:
            if var not in ds:
                spatial[var].append(np.full((len(lats), len(lons)), np.nan))
                ts_vals[var].append(np.nan)
                continue
            da  = ds[var].sel(lat=slice(lat_min, lat_max),
                               lon=slice(lon_min, lon_max))
            arr = da.mean(dim="orbit_pass", skipna=True).values   # (lat, lon)
            spatial[var].append(arr)
            ts_vals[var].append(float(np.nanmean(arr)))
        dates.append(date)

dates    = pd.DatetimeIndex(dates)
n_frames = len(dates)
print(f"Loaded {n_frames} months.\n")

# ── Phase 2: animation factory ─────────────────────────────────────────────
def animate_variable(var):
    m  = meta[var]
    sc, of = m["scale"], m["offset"]

    # scaled spatial stacks and time series
    maps  = [a * sc + of for a in spatial[var]]
    y_all = np.array(ts_vals[var]) * sc + of
    roll  = (pd.Series(y_all, index=dates)
               .rolling(12, center=True, min_periods=6)
               .mean().values)

    # colour limits: 2nd–98th percentile across all frames
    all_px = np.concatenate([a.ravel() for a in maps])
    vmin   = np.nanpercentile(all_px,  2)
    vmax   = np.nanpercentile(all_px, 98)

    ts_ymin = np.nanmin(y_all)
    ts_ymax = np.nanmax(y_all)
    ts_pad  = (ts_ymax - ts_ymin) * 0.05

    proj = ccrs.PlateCarree()
    fig  = plt.figure(figsize=(11, 8.5), facecolor="white")
    gs   = fig.add_gridspec(2, 1, height_ratios=[1.6, 1.0], hspace=0.42)

    # ── map panel ─────────────────────────────────────────────────────────
    ax_map = fig.add_subplot(gs[0], projection=proj)
    ax_map.set_extent([lon_min, lon_max, lat_min, lat_max], crs=proj)
    ax_map.coastlines(resolution="50m", linewidth=0.9)
    ax_map.add_feature(cfeature.BORDERS, linewidth=0.5, linestyle="--")
    ax_map.add_feature(cfeature.STATES,  linewidth=0.3)
    gl = ax_map.gridlines(draw_labels=True, linewidth=0.4, linestyle=":")
    gl.top_labels = gl.right_labels = False

    pcm = ax_map.pcolormesh(
        lons, lats, maps[0], transform=proj,
        cmap=m["cmap"], vmin=vmin, vmax=vmax, shading="nearest",
    )
    cb = plt.colorbar(pcm, ax=ax_map, orientation="vertical", pad=0.03, shrink=0.88)
    cb.set_label(m["disp_unit"], fontsize=9)
    map_ttl = ax_map.set_title(
        f"{m['long_name']}  |  {dates[0].strftime('%B %Y')}", fontsize=10)

    # ── time-series panel ─────────────────────────────────────────────────
    ax_ts = fig.add_subplot(gs[1])
    ax_ts.set_xlim(dates[0], dates[-1])
    ax_ts.set_ylim(ts_ymin - ts_pad, ts_ymax + ts_pad)
    ax_ts.set_ylabel(m["disp_unit"], fontsize=9)
    ax_ts.set_xlabel("Date", fontsize=9)
    ax_ts.xaxis.set_major_locator(mdates.YearLocator(2))
    ax_ts.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
    ax_ts.xaxis.set_minor_locator(mdates.YearLocator())
    ax_ts.yaxis.set_minor_locator(AutoMinorLocator(4))
    ax_ts.grid(axis="y", which="major", lw=0.5, alpha=0.55)
    ax_ts.grid(axis="y", which="minor", lw=0.2, alpha=0.3)
    ax_ts.grid(axis="x", which="minor", lw=0.2, alpha=0.25)

    ts_ln,   = ax_ts.plot([], [], lw=1.4, color=m["line_color"], zorder=3)
    roll_ln, = ax_ts.plot([], [], lw=2.0, color="k", alpha=0.45,
                          linestyle="--", label="12-month rolling mean", zorder=4)
    # vertical marker for current frame
    vline    = ax_ts.axvline(dates[0], color="gray", lw=1.0, alpha=0.75, zorder=5)
    ax_ts.legend(fontsize=8, loc="upper left")

    # mutable container so fill_between can be replaced each frame
    fill = [None]

    fig.suptitle(f"AIRS CLIMCAPS  –  {m['long_name']}  "
                 f"(lon {lon_min}°–{lon_max}°, lat {lat_min}°–{lat_max}°)",
                 fontsize=11, y=1.00)

    def update(i):
        # update map
        pcm.set_array(maps[i].ravel())
        map_ttl.set_text(f"{m['long_name']}  |  {dates[i].strftime('%B %Y')}")

        # update growing time series
        t_now = dates[:i + 1]
        y_now = y_all[:i + 1]
        ts_ln.set_data(t_now, y_now)
        roll_ln.set_data(t_now, roll[:i + 1])
        vline.set_xdata([dates[i], dates[i]])

        # replace fill_between with updated polygon
        if fill[0] is not None:
            fill[0].remove()
        fill[0] = ax_ts.fill_between(
            t_now, y_now, ts_ymin - ts_pad,
            alpha=0.45, color=m["fill_color"], zorder=2,
        )

        if (i + 1) % 60 == 0 or i == n_frames - 1:
            print(f"    frame {i + 1}/{n_frames}")

    anim = mpl_anim.FuncAnimation(
        fig, update, frames=n_frames, interval=150, blit=False)

    out_mp4 = os.path.join(SAVE_DIR, f"animation_{var}.mp4")
    out_gif = os.path.join(SAVE_DIR, f"animation_{var}.gif")
    saved   = None

    try:
        writer = mpl_anim.FFMpegWriter(
            fps=6, bitrate=2000,
            extra_args=["-vcodec", "libx264", "-pix_fmt", "yuv420p"],
        )
        print(f"  Saving MP4 …")
        anim.save(out_mp4, writer=writer, dpi=110)
        saved = out_mp4
    except Exception as exc:
        print(f"  FFMpeg failed ({exc})\n  Falling back to GIF …")
        writer = mpl_anim.PillowWriter(fps=6)
        anim.save(out_gif, writer=writer, dpi=90)
        saved = out_gif

    plt.close(fig)
    print(f"  → Saved: {saved}\n")
    return saved

# ── Phase 3: run for every variable ────────────────────────────────────────
saved_paths = {}
for var in VARIABLES:
    print(f"── Animating: {meta[var]['long_name']} ──")
    saved_paths[var] = animate_variable(var)

print("\nAll animations complete:")
for var, p in saved_paths.items():
    print(f"  {var:25s}  →  {p}")
```

```python
import os, re, glob
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.animation as mpl_anim
from matplotlib.ticker import AutoMinorLocator, FuncFormatter
import cartopy.crs as ccrs
import cartopy.feature as cfeature

matplotlib.rcParams["animation.ffmpeg_path"] = (
    r"C:\Users\ACER\anaconda3\envs\gis-env\Library\bin\ffmpeg.exe"
)

# ── region (same bounding box throughout) ────────────────────────────────────
lon_min, lon_max = -107.5, -85.0
lat_min, lat_max =   24.5,  37.5
DATA_DIR = r"D:\COCCON\Work\Data\SNDRAQIL3SMCCP_2"
SAVE_DIR = r"D:\COCCON\Work\Figures"
os.makedirs(SAVE_DIR, exist_ok=True)

REGION_STR = (
    f"lon {lon_min}°–{lon_max}°, lat {lat_min}°–{lat_max}°  |  "
    "spatial mean  |  Sep 2002 – Dec 2025"
)

# ── metadata ───────────────────────────────────────────────────────────────────────────
# ch4 → purple; surf_air_temp → orange-red  (both overwrite existing outputs)
# surf_temp and air_temp are new outputs

meta_2d = {
    "ch4_mmr_midtrop": dict(
        long_name="CH₄ Mid-Troposphere MMR", disp_unit="ppb",
        scale=1e9, offset=0.0,
        line_color="#6c3483",   # deep purple
        fill_color="#d7bde2",   # soft lavender
        cmap="RdPu",
    ),
    "surf_air_temp": dict(
        long_name="Near-Surface Air Temperature", disp_unit="°C",
        scale=1.0, offset=-273.15,
        line_color="#b7410e",   # rust-orange
        fill_color="#fde3c0",   # warm apricot
        cmap="YlOrRd",
    ),
    "surf_temp": dict(
        long_name="Surface Skin Temperature", disp_unit="°C",
        scale=1.0, offset=-273.15,
        line_color="#922b21",   # deep brick-red
        fill_color="#fadbd8",   # blush-pink
        cmap="Reds",
    ),
}

meta_air = dict(
    long_name="Air Temperature",
    disp_unit="°C",
    scale=1.0, offset=-273.15,
    line_color="#154360",   # dark navy
    fill_color="#d6eaf8",   # pale steel-blue
    cmap="RdYlBu_r",
)

REG_VARS = list(meta_2d.keys())

# ── Phase 1 – load all files once ──────────────────────────────────────────────
files    = sorted(glob.glob(os.path.join(DATA_DIR, '*.nc')))
date_pat = re.compile(r"\.(\d{8})\.")

spatial_2d  = {v: [] for v in REG_VARS}
ts_vals_2d  = {v: [] for v in REG_VARS}

air_map_data = []
air_profile  = []
ts_air_temp  = []
air_pres_hpa = None

dates = []
lats = lons = None

print(f"Loading {len(files)} files …")
for f in files:
    hit = date_pat.search(os.path.basename(f))
    if not hit:
        continue
    date = pd.Timestamp(hit.group(1))
    if date.year > 2025:
        continue
    with xr.open_dataset(f, decode_timedelta=False) as ds:
        if lats is None:
            lats         = ds["lat"].sel(lat=slice(lat_min, lat_max)).values
            lons         = ds["lon"].sel(lon=slice(lon_min, lon_max)).values
            air_pres_hpa = ds["air_pres"].values / 100.0

        for var in REG_VARS:
            if var not in ds:
                spatial_2d[var].append(np.full((len(lats), len(lons)), np.nan))
                ts_vals_2d[var].append(np.nan)
                continue
            da  = ds[var].sel(lat=slice(lat_min, lat_max),
                               lon=slice(lon_min, lon_max))
            arr = da.mean(dim="orbit_pass", skipna=True).values
            spatial_2d[var].append(arr)
            ts_vals_2d[var].append(float(np.nanmean(arr)))

        n_pres = len(air_pres_hpa) if air_pres_hpa is not None else 100
        if "air_temp" in ds:
            at   = ds["air_temp"].sel(lat=slice(lat_min, lat_max),
                                       lon=slice(lon_min, lon_max))
            at3d = at.mean(dim="orbit_pass", skipna=True).values
            air_map_data.append(np.nanmean(at3d, axis=0))
            air_profile.append(np.nanmean(at3d, axis=(1, 2)))
            ts_air_temp.append(float(np.nanmean(at3d)))
        else:
            air_map_data.append(np.full((len(lats), len(lons)), np.nan))
            air_profile.append(np.full(n_pres, np.nan))
            ts_air_temp.append(np.nan)

        dates.append(date)

dates    = pd.DatetimeIndex(dates)
n_frames = len(dates)
print(f"Loaded {n_frames} months.\n")


# ── Phase 2 – helpers ───────────────────────────────────────────────────────────────────

def _save_mp4_or_gif(anim, var_name):
    out_mp4 = os.path.join(SAVE_DIR, f'animation_{var_name}.mp4')
    out_gif = os.path.join(SAVE_DIR, f'animation_{var_name}.gif')
    try:
        writer = mpl_anim.FFMpegWriter(
            fps=6, bitrate=2000,
            extra_args=["-vcodec", "libx264", "-pix_fmt", "yuv420p"],
        )
        anim.save(out_mp4, writer=writer, dpi=110)
        print(f"  Animation → {out_mp4}")
    except Exception as exc:
        print(f"  FFMpeg failed ({exc}), falling back to GIF …")
        anim.save(out_gif, writer=mpl_anim.PillowWriter(fps=6), dpi=90)
        print(f"  Animation → {out_gif}")


def make_ts_plot(var_name, m, y_all, title_suffix=""):
    roll = (pd.Series(y_all, index=dates)
              .rolling(12, center=True, min_periods=6).mean())
    fig, ax = plt.subplots(figsize=(14, 4.5))
    ax.plot(dates, y_all, lw=1.4, color=m['line_color'], zorder=3)
    ax.fill_between(dates, y_all, np.nanmin(y_all),
                    alpha=0.45, color=m['fill_color'], zorder=2)
    ax.plot(dates, roll, lw=2.0, color='k', alpha=0.45,
            linestyle='--', label='12-month rolling mean')
    ax.legend(fontsize=9, loc='upper left')
    ax.set_title(
        f"AIRS CLIMCAPS  –  {m['long_name']}{title_suffix}\n{REGION_STR}",
        fontsize=11, pad=8,
    )
    ax.set_ylabel(m['disp_unit'], fontsize=10)
    ax.set_xlabel('Date', fontsize=10)
    ax.xaxis.set_major_locator(mdates.YearLocator(2))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    ax.xaxis.set_minor_locator(mdates.YearLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator(4))
    ax.grid(axis='y', which='major', lw=0.5, alpha=0.55)
    ax.grid(axis='y', which='minor', lw=0.2, alpha=0.3)
    ax.grid(axis='x', which='minor', lw=0.2, alpha=0.25)
    ax.set_xlim(dates[0], dates[-1])
    fig.tight_layout()
    path = os.path.join(SAVE_DIR, f'timeseries_{var_name}.png')
    fig.savefig(path, dpi=150, bbox_inches='tight')
    plt.close(fig)
    print(f"  Plot    → {path}")


def animate_2d(var_name, m, maps, y_all):
    roll    = (pd.Series(y_all, index=dates)
                 .rolling(12, center=True, min_periods=6).mean().values)
    all_px  = np.concatenate([a.ravel() for a in maps])
    vmin    = np.nanpercentile(all_px,  2)
    vmax    = np.nanpercentile(all_px, 98)
    ts_ymin = np.nanmin(y_all);  ts_ymax = np.nanmax(y_all)
    ts_pad  = (ts_ymax - ts_ymin) * 0.05

    proj = ccrs.PlateCarree()
    fig  = plt.figure(figsize=(11, 8.5), facecolor='white')
    gs   = fig.add_gridspec(2, 1, height_ratios=[1.6, 1.0], hspace=0.42)

    ax_map = fig.add_subplot(gs[0], projection=proj)
    ax_map.set_extent([lon_min, lon_max, lat_min, lat_max], crs=proj)
    ax_map.coastlines(resolution='50m', linewidth=0.9)
    ax_map.add_feature(cfeature.BORDERS, linewidth=0.5, linestyle='--')
    ax_map.add_feature(cfeature.STATES,  linewidth=0.3)
    gl = ax_map.gridlines(draw_labels=True, linewidth=0.4, linestyle=':')
    gl.top_labels = gl.right_labels = False

    pcm = ax_map.pcolormesh(lons, lats, maps[0], transform=proj,
                            cmap=m['cmap'], vmin=vmin, vmax=vmax, shading='nearest')
    cb  = plt.colorbar(pcm, ax=ax_map, orientation='vertical', pad=0.03, shrink=0.88)
    cb.set_label(m['disp_unit'], fontsize=9)
    map_ttl = ax_map.set_title(
        f"{m['long_name']}  |  {dates[0].strftime('%B %Y')}", fontsize=10)

    ax_ts = fig.add_subplot(gs[1])
    ax_ts.set_xlim(dates[0], dates[-1])
    ax_ts.set_ylim(ts_ymin - ts_pad, ts_ymax + ts_pad)
    ax_ts.set_ylabel(m['disp_unit'], fontsize=9)
    ax_ts.set_xlabel('Date', fontsize=9)
    ax_ts.xaxis.set_major_locator(mdates.YearLocator(2))
    ax_ts.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    ax_ts.xaxis.set_minor_locator(mdates.YearLocator())
    ax_ts.yaxis.set_minor_locator(AutoMinorLocator(4))
    ax_ts.grid(axis='y', which='major', lw=0.5, alpha=0.55)
    ax_ts.grid(axis='y', which='minor', lw=0.2, alpha=0.3)
    ax_ts.grid(axis='x', which='minor', lw=0.2, alpha=0.25)

    ts_ln,   = ax_ts.plot([], [], lw=1.4, color=m['line_color'], zorder=3)
    roll_ln, = ax_ts.plot([], [], lw=2.0, color='k', alpha=0.45,
                          linestyle='--', label='12-month rolling mean', zorder=4)
    vline    = ax_ts.axvline(dates[0], color='gray', lw=1.0, alpha=0.75, zorder=5)
    fill     = [None]
    ax_ts.legend(fontsize=8, loc='upper left')
    fig.suptitle(
        f"AIRS CLIMCAPS  –  {m['long_name']}  "
        f"(lon {lon_min}°–{lon_max}°, lat {lat_min}°–{lat_max}°)",
        fontsize=11, y=1.00,
    )

    def update(i):
        pcm.set_array(maps[i].ravel())
        map_ttl.set_text(f"{m['long_name']}  |  {dates[i].strftime('%B %Y')}")
        t_now = dates[:i+1];  y_now = y_all[:i+1]
        ts_ln.set_data(t_now, y_now)
        roll_ln.set_data(t_now, roll[:i+1])
        vline.set_xdata([dates[i], dates[i]])
        if fill[0] is not None:
            fill[0].remove()
        fill[0] = ax_ts.fill_between(
            t_now, y_now, ts_ymin - ts_pad,
            alpha=0.45, color=m['fill_color'], zorder=2,
        )
        if (i + 1) % 60 == 0 or i == n_frames - 1:
            print(f'    frame {i+1}/{n_frames}')

    anim = mpl_anim.FuncAnimation(fig, update, frames=n_frames, interval=150, blit=False)
    _save_mp4_or_gif(anim, var_name)
    plt.close(fig)


def animate_air_temp():
    m  = meta_air
    of = m['offset']

    maps_c     = [a + of for a in air_map_data]
    profiles_c = [p + of for p in air_profile]
    y_all      = np.array(ts_air_temp) + of
    roll       = (pd.Series(y_all, index=dates)
                    .rolling(12, center=True, min_periods=6).mean().values)

    all_px     = np.concatenate([a.ravel() for a in maps_c])
    vmin       = np.nanpercentile(all_px,  2)
    vmax       = np.nanpercentile(all_px, 98)

    all_pr     = np.concatenate(profiles_c)
    t_lo       = np.nanpercentile(all_pr,  1)
    t_hi       = np.nanpercentile(all_pr, 99)
    clim_prof  = np.nanmean(np.vstack(profiles_c), axis=0)

    p_bot      = float(np.nanmax(air_pres_hpa))
    p_top      = float(np.nanmin(air_pres_hpa))

    ts_ymin    = np.nanmin(y_all);  ts_ymax = np.nanmax(y_all)
    ts_pad     = (ts_ymax - ts_ymin) * 0.05

    proj = ccrs.PlateCarree()
    fig  = plt.figure(figsize=(13, 9.5), facecolor='white')
    gs   = fig.add_gridspec(
        2, 2,
        height_ratios=[1.55, 1.0],
        width_ratios=[2.4,  1.0],
        hspace=0.42, wspace=0.32,
    )
    ax_map  = fig.add_subplot(gs[0, 0], projection=proj)
    ax_prof = fig.add_subplot(gs[0, 1])
    ax_ts   = fig.add_subplot(gs[1, :])

    ax_map.set_extent([lon_min, lon_max, lat_min, lat_max], crs=proj)
    ax_map.coastlines(resolution='50m', linewidth=0.9)
    ax_map.add_feature(cfeature.BORDERS, linewidth=0.5, linestyle='--')
    ax_map.add_feature(cfeature.STATES,  linewidth=0.3)
    gl = ax_map.gridlines(draw_labels=True, linewidth=0.4, linestyle=':')
    gl.top_labels = gl.right_labels = False

    pcm = ax_map.pcolormesh(lons, lats, maps_c[0], transform=proj,
                            cmap=m['cmap'], vmin=vmin, vmax=vmax, shading='nearest')
    cb  = plt.colorbar(pcm, ax=ax_map, orientation='vertical', pad=0.03, shrink=0.88)
    cb.set_label('°C', fontsize=9)
    map_ttl = ax_map.set_title(
        f"Air Temp (column mean)  |  {dates[0].strftime('%B %Y')}", fontsize=10)

    ax_prof.set_facecolor('#eaf2f8')
    for sp in ax_prof.spines.values():
        sp.set_linewidth(1.8);  sp.set_color('#2c3e50')

    ax_prof.set_yscale('log')
    ax_prof.set_ylim(p_bot, p_top)
    ax_prof.set_xlim(t_lo - 3, t_hi + 3)
    ax_prof.set_xlabel('Temperature (°C)', fontsize=8)
    ax_prof.set_ylabel('Pressure (hPa)',   fontsize=8)
    ax_prof.tick_params(labelsize=7)
    ax_prof.set_yticks([0.1, 1, 10, 100, 1000])
    ax_prof.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y:g}'))
    ax_prof.grid(which='both', lw=0.35, alpha=0.55, color='white')
    ax_prof.set_title('Vertical T Profile\n(region spatial mean)', fontsize=8.5, pad=4)

    ax_prof.plot(clim_prof, air_pres_hpa,
                 lw=1.2, color='#7f8c8d', alpha=0.65,
                 linestyle=':', label='climatology', zorder=2)
    prof_ln, = ax_prof.plot(profiles_c[0], air_pres_hpa,
                             lw=2.0, color='#e74c3c', zorder=3, label='current month')
    ax_prof.legend(fontsize=7, loc='lower left', framealpha=0.7)

    ax_ts.set_xlim(dates[0], dates[-1])
    ax_ts.set_ylim(ts_ymin - ts_pad, ts_ymax + ts_pad)
    ax_ts.set_ylabel('°C  (column mean)', fontsize=9)
    ax_ts.set_xlabel('Date', fontsize=9)
    ax_ts.xaxis.set_major_locator(mdates.YearLocator(2))
    ax_ts.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    ax_ts.xaxis.set_minor_locator(mdates.YearLocator())
    ax_ts.yaxis.set_minor_locator(AutoMinorLocator(4))
    ax_ts.grid(axis='y', which='major', lw=0.5, alpha=0.55)
    ax_ts.grid(axis='y', which='minor', lw=0.2, alpha=0.3)
    ax_ts.grid(axis='x', which='minor', lw=0.2, alpha=0.25)

    ts_ln,   = ax_ts.plot([], [], lw=1.4, color=m['line_color'], zorder=3)
    roll_ln, = ax_ts.plot([], [], lw=2.0, color='k', alpha=0.45,
                          linestyle='--', label='12-month rolling mean', zorder=4)
    vline    = ax_ts.axvline(dates[0], color='gray', lw=1.0, alpha=0.75, zorder=5)
    fill     = [None]
    ax_ts.legend(fontsize=8, loc='upper left')
    fig.suptitle(
        f"AIRS CLIMCAPS  –  Air Temperature  "
        f"(lon {lon_min}°–{lon_max}°, lat {lat_min}°–{lat_max}°)",
        fontsize=11, y=1.00,
    )

    def update(i):
        pcm.set_array(maps_c[i].ravel())
        map_ttl.set_text(
            f"Air Temp (column mean)  |  {dates[i].strftime('%B %Y')}")
        prof_ln.set_data(profiles_c[i], air_pres_hpa)
        t_now = dates[:i+1];  y_now = y_all[:i+1]
        ts_ln.set_data(t_now, y_now)
        roll_ln.set_data(t_now, roll[:i+1])
        vline.set_xdata([dates[i], dates[i]])
        if fill[0] is not None:
            fill[0].remove()
        fill[0] = ax_ts.fill_between(
            t_now, y_now, ts_ymin - ts_pad,
            alpha=0.45, color=m['fill_color'], zorder=2,
        )
        if (i + 1) % 60 == 0 or i == n_frames - 1:
            print(f'    frame {i+1}/{n_frames}')

    anim = mpl_anim.FuncAnimation(fig, update, frames=n_frames, interval=150, blit=False)
    _save_mp4_or_gif(anim, 'air_temp')
    plt.close(fig)


# ── Phase 3 – generate all outputs ─────────────────────────────────────────────────────

print('=== Static time-series plots ===')
for var in REG_VARS:
    m     = meta_2d[var]
    y_all = np.array(ts_vals_2d[var]) * m['scale'] + m['offset']
    print(f"── {m['long_name']}")
    make_ts_plot(var, m, y_all)

print(f"── {meta_air['long_name']} (column mean)")
make_ts_plot(
    'air_temp', meta_air,
    np.array(ts_air_temp) + meta_air['offset'],
    title_suffix=' (column mean, 100 pressure levels)',
)

print('\n=== Animations ===')
for var in REG_VARS:
    m      = meta_2d[var]
    sc, of = m['scale'], m['offset']
    print(f"── {m['long_name']}")
    animate_2d(var, m,
               [a * sc + of for a in spatial_2d[var]],
               np.array(ts_vals_2d[var]) * sc + of)

print('── Air Temperature (map + vertical profile + time series)')
animate_air_temp()

print('\nAll done!')
```
